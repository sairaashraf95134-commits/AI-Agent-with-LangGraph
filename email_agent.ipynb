```python
import yagmail
yag = yagmail.SMTP(
    "sairaashraf95134@gmail.com",
    "hxfy cmab jzwn dagh"
)

print("Yagmail Connected Successfully!")
```

```python
from langchain.tools import tool
@tool
def send_email_tool(recipient: str, subject: str, body: str) -> str:
    """
    Send email to recipient with subject and body.
    """

    yag.send(
        to=recipient,
        subject=subject,
        contents=body
    )

    return "Email sent successfully!"
```

```python
from langchain.agents import create_agent

from langchain_groq import ChatGroq

llm = ChatGroq(
    model="openai/gpt-oss-120b",
    api_key="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"
)

print("Groq Model Connected!")

agent = create_agent(
    model=llm,
    tools=[send_email_tool],
    system_prompt="You are an intelligent email sending agent."
)

print("Agent Created Successfully!")
```

```python
user_request = input("Enter your email request: ")
response = agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": user_request
            }
        ]
    }
)

print(response)
```
